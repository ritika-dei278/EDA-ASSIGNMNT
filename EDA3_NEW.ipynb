```python
df['discounted_price'] = df['discounted_price'].replace(to_replace={'₹':'',',':''},regex=True)
```

```python
df.discounted_price = df.discounted_price.astype(float)
```

```python
df.discounted_price.dtype
```

```python
df.dtypes
```

```python
df['actual_price'] = df['actual_price'].replace(to_replace={'₹':'',',':''},regex=True)
```

```python
df.actual_price = df.actual_price.astype(float)
```

```python
df.actual_price.dtype
```

```python
df.dtypes
```

```python
df.discount_percentage = df.discount_percentage.str.replace('%','')
```

```python
df.discount_percentage = df.discount_percentage.astype(float)
```

```python
#3  What is the distribution of discounted prices vs. actual prices

plt.figure(figsize=(10, 6))
sns.scatterplot(x='actual_price', y='discounted_price', data=df)
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```

```python
#as the actual price rises the discounted price also rises.
```

```python
#7  What are the most popular product reviews

popular_reviews = df.sort_values(by=['rating', 'rating_count'], ascending=False)

top_10_popular = popular_reviews[['product_name', 'rating', 'rating_count']].head(10)

print("Most Popular Product Reviews (based on rating and rating count):")
print(top_10_popular)
```

```python
#8  What is the correlation between discounted_price and rating
columns = ['discounted_price','rating']
sns.heatmap(df[columns].corr(), annot = True, cmap = 'coolwarm')
```
