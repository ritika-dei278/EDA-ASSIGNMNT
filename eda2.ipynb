```python
#1. What is the average selling price of cars ?for each dealer, and how does it compare across different dealer?
dealer_avg_price = df.groupby("Dealer_Name")['Price ($)'].mean().sort_values(ascending=False)
dealer_avg_price
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=dealer_avg_price.index, y=dealer_avg_price.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Dealer Name')
plt.ylabel('Average Selling Price ($)')
plt.title('Average Selling Price of Cars by Dealer')
plt.tight_layout()
plt.show()
```

INSIGHTS :
The Highest avg. SP is by U-HAUL CO. - $28769.919006.

The average selling price does not vary greatly among different dealers. It ranges between $25000 -30000 across each dealer.

```python
#2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
#trends

company_price_variation = df.groupby('Company')['Price ($)'].std().sort_values(ascending=False)

highest_variation_company = company_price_variation.index[0]

print(f"The car brand with the highest price variation is: {highest_variation_company}")
print(company_price_variation)
```

The comapnies with high price variation like lincoln, saab, etc. tells us that the cars sold by thse companies differs significantly in their price range.

```python
#3, What is the distribution of car prices for each transmission type, and how do the interquartile ranges
#compare
df.columns
```

```python
transmission_stats = df.groupby('Transmission')['Price ($)'].describe()
print(transmission_stats)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Transmission', y='Price ($)', data=df)
plt.title('Distribution of Car Prices by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Extracting IQR values from the descriptive statistics
automatic_iqr = transmission_stats.loc['Auto', '75%'] - transmission_stats.loc['Auto', '25%']
manual_iqr = transmission_stats.loc['Manual', '75%'] - transmission_stats.loc['Manual', '25%']

print(f"IQR for Automatic Transmission: {automatic_iqr}")
print(f"IQR for Manual Transmission: {manual_iqr}")
```

```python
#12  Is there any seasonal trend in car sales based on the date of sale
df['month_of_carsale'] = pd.DatetimeIndex(df['Date']).month
df['month_of_carsale']
```
