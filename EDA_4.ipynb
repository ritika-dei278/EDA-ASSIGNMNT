```python
#2  What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram
sns.histplot(x =df['Popularity'], kde = True, color = 'purple')
plt.title('Distribution of popularity among the tracks in the dataset')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
#the histogram is left-skewed meaning that most of the tracks have less popularity scores
```

```python
#3  Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot
sns.scatterplot(x = df['Popularity'], y = df['Duration (ms)'], color = 'red')
plt.title('Relationship between the popularity and the duration of tracks')
plt.xlabel('Popularity')
plt.ylabel('Duration_ms')
plt.show()
```

```python
#4 Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using
#a countplot
plt.figure(figsize = (17,8))
sns.countplot(x = df['Artist'])
plt.xticks(rotation = 90)
plt.title('Count of tracks for each artist')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.show()
```

```python
#DRAKE has th highest tracks in the dataset.
```

```python

#5  What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
least_popular_tracks = df.sort_values(by=['Popularity'], ascending=True).head(5)

# Display the artist name and track name for the least popular tracks
for index, row in least_popular_tracks.iterrows():
    print(f"Artist: {row['Artist']}, Track: {row['Track Name']}, Popularity: {row['Popularity']}")
```

```python
#6 Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and
#display the average popularity for each artist
avg_highest_pop = df.groupby('Artist')['Popularity'].mean().sort_values(ascending = False).head()
```

```python
sns.barplot(x = avg_highest_pop.index, y = avg_highest_pop.values)
```

```python
#INSIGHTS > artist named casso has the highesr average popularity followed by Trueno
```

```python
#7  For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist
artist_popularity = df.loc[df.groupby('Artist')['Popularity'].idxmax()]  # Get the rows with maximum popularity for each artist
artist_popularity = artist_popularity.sort_values(by=['Popularity'], ascending=False).head(5) # Sort by popularity and get the top 5

for index, row in artist_popularity.iterrows():
    print(f"Artist: {row['Artist']}, Track: {row['Track Name']}, Popularity: {row['Popularity']}")
```

```python
#8  Visualize relationships between multiple numerical variables simultaneously using a pair plot
numerical_vars = ['Popularity', 'Duration (ms)'] #sorting numerical columns
sns.pairplot(df[numerical_vars])
plt.show()
```

```python
#9 Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
#violin plot

plt.figure(figsize=(16, 6))  # Adjust figure size as needed
sns.boxplot(x='Artist', y='Duration (ms)', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Duration of Tracks Across Different Artists')
plt.show()
```

```python
#10 How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
#violin plot.

plt.figure(figsize=(16, 6))  # Adjust figure size as needed
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Distribution of Track Popularity Across Different Artists')
plt.show()
```
